```python
import calendar
import datetime
import json
import os
import time

REMINDER_FILE = "reminders.json"

# Load reminders from file
def load_reminders():
    if os.path.exists(REMINDER_FILE):
        with open(REMINDER_FILE, "r") as file:
            return json.load(file)
    return {}

# Save reminders to file
def save_reminders(reminders):
    with open(REMINDER_FILE, "w") as file:
        json.dump(reminders, file, indent=4)

# Show calendar
def show_calendar(year, month):
    print("\n" + calendar.month(year, month))

# Add a reminder with date and time
def add_reminder(reminders):
    date_str = input("Enter date (YYYY-MM-DD): ")
    time_str = input("Enter time (HH:MM, 24-hour): ")

    # Validate datetime
    try:
        datetime_obj = datetime.datetime.strptime(f"{date_str} {time_str}", "%Y-%m-%d %H:%M")
    except ValueError:
        print("Invalid date or time format.")
        return

    reminder_text = input("Enter reminder: ")
    datetime_key = datetime_obj.strftime("%Y-%m-%d %H:%M")

    if datetime_key in reminders:
        reminders[datetime_key].append(reminder_text)
    else:
        reminders[datetime_key] = [reminder_text]

    save_reminders(reminders)
    print("Reminder added!")

# View reminders for a date
def view_reminders(reminders):
    date_str = input("Enter date to view reminders (YYYY-MM-DD): ")
    print(f"\nReminders on {date_str}:")
    found = False
    for key in sorted(reminders.keys()):
        if key.startswith(date_str):
            found = True
            print(f"{key} ->")
            for r in reminders[key]:
                print(f"  - {r}")
    if not found:
        print("No reminders found.")

# Check for current reminder
def check_current_reminders(reminders):
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M")
    if now in reminders:
        print("\nðŸ”” You have reminders now!")
        for r in reminders[now]:
            print(f"  - {r}")

# Menu
def main():
    reminders = load_reminders()

    while True:
        check_current_reminders(reminders)
        print("\n=== Calendar & Reminder App ===")
        print("1. Show calendar")
        print("2. Add reminder (with time)")
        print("3. View reminders for a date")
        print("4. Exit")

        choice = input("Select option: ")

        if choice == '1':
            try:
                year = int(input("Enter year (e.g., 2025): "))
                month = int(input("Enter month (1-12): "))
                show_calendar(year, month)
            except ValueError:
                print("Invalid year or month.")
        elif choice == '2':
            add_reminder(reminders)
        elif choice == '3':
            view_reminders(reminders)
        elif choice == '4':
            print("Goodbye!")
            break
        else:
            print("Invalid option.")

        # Wait a few seconds before re-displaying
        time.sleep(2)

if __name__ == "__main__":
    main()
```
